## Imoveis Residenciais
<b> Analisando Series </b>

```python
#Analisando a série Tipo de Imovel
tipo_imovel = dados['Tipo']
tipo_imovel
```

```python
#removendo as duplicações
#lembrando que, para apenas vizualizar os dados sem alterar a variável, dentro do parenteses do metodo drop_duplicate deve estar em braco.
tipo_imovel.drop_duplicates()
```

```python
# filtrando os tipos de imoveis que queremos no dataframe:
residenciais = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',
 'Chácara',
 'Sítio']
```

```python
#através da matriz booleana, teremos os valores iguais a True na para os tipos de imóveis que separamos na lista "residenciais" 
selecao = dados['Tipo'].isin(residenciais)
selecao
```

```python
# com isso, geramos um novo dataframe, agora contendo apenas os imoveis residenciais que nos interessam.
dados_residencial = dados[selecao]
dados_residencial
```

```python
#formatacação do index
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

```python
#exportando a fonte de dados após o filtro de tipo de Imóvel
dados_residencial.to_csv(r'C:\Users\gpere\Desktop\Cursos\Alura\Data Science\Python Pandas\dataframe\aluguel_residencial.csv', sep =';', index = False)
```
